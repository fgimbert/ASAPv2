# Automatic Surface Adsorption Predictor (ASAP) TEST !

The objective to ASAP notebook is to automatize the building and optimization of surface with adsorption structures (atom, molecule) by Bayesian Optimization. The concept idea is to use 'Building Blocks' to accelerate the optimization by keeping low the number of free parameters. 

**See paper: Bayesian inference of atomistic structure in functional materials (npj Computational Materials, 2019, 35)**

To do this, ASAP works in 3 modules:

1. the search module
2. the building module
3. the optimization module

All these modules are included in a interface inside this notebook realized with ipywidgets.

### 1. Search module

This module allows the user to look for any materials inside Materials Project Database (in a second time, Nomad will be added) with desired compositions or properties. The selected material can be used as material for surface system.
If the chemical search is chosen, all the materials with the specified chemical elements will be imported to allow the constrcution of phase diagram. 

It is also possible to look for molecule on Materials Project database.

### 2. Building module

This module allows the user to build a surface with adsorption structure (atom or molecule) with specific unit cell size, thickness for slab. The inner geometry of the adsorption structure is fixed as building block but the geometry surface-adsorbant is let free with specific free parameters extracted automatically (height, orientation, position). These parameters will be optimized in the third module. 

**Important** For now, only the molecules available inside ase database are available. A function to transform Materials Project molecule to ASE Atoms object will be added in the future.
The list of molecules available can be found here : https://wiki.fysik.dtu.dk/ase/ase/build/build.html#molecules


### 3. Optimization module

This module perform the bayesian optimization of the adsorption structure builded in the Building module.

Pure Python EMT calculator and VASP can be currently used (in future, OpenMx will be added).

**VASP configuration**
The potentials for VASP need to be a directory called potvasp and inside potpaw and potpaw_PBE repertory.
THe jobfile needs to be modified for your own cluster (inside configuration tabular). Keep JOBNAME as job name ! (very important)

**For EMT:**  
**Right now, the only supported elements are: H, C, N, O, Al, Ni, Cu, Pd, Ag, Pt and Au.** The EMT parameters for the metals are quite realistic for many purposes, whereas the H, C, N and O parameters are just for fun!

```python
# Set up some imports that we will need
from pymatgen import Lattice, Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from IPython.display import Image, display

%matplotlib inline
```

```python
# Create CsCl structure
a = 4.209 #Angstrom
latt = Lattice.cubic(a)
structure = Structure(latt, ["Cs", "Cl"], [[0, 0, 0], [0.5, 0.5, 0.5]])

c = XRDCalculator()
c.show_plot(structure)
```

```python

"""
create nanoparticle using wulff construction
"""

from pymatgen.symmetry.analyzer import SpacegroupAnalyzer

from mpinterfaces import get_struct_from_mp
from mpinterfaces.nanoparticle import Nanoparticle

# -----------------------------------
# nanoparticle specifications
# -----------------------------------
# max radius in angstroms
rmax = 15
# surface families to be chopped off
hkl_family = [(1, 0, 0), (1, 1, 1)]
# surfac energies could be in any units, will be normalized
surface_energies = [28, 25]

# -----------------------------------
# initial structure
# -----------------------------------
# caution: set the structure wrt which the the miller indices are
# specified. use your own key
structure = get_struct_from_mp('PbS', MAPI_KEY="")
# primitive ---> conventional cell
sa = SpacegroupAnalyzer(structure)
structure_conventional = sa.get_conventional_standard_structure()

# -----------------------------------
# create nanoparticle
# -----------------------------------
nanoparticle = Nanoparticle(structure_conventional, rmax=rmax,
                            hkl_family=hkl_family,
                            surface_energies=surface_energies)
nanoparticle.create()
nanoparticle.to(fmt='xyz', filename='nanoparticle.xyz')
```
